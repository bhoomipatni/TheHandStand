# Exploration Notebook

This notebook is intended for exploratory data analysis and visualization of the dataset used in the sign skeleton pipeline project. We will analyze the downloaded videos, extracted keypoints, and skeleton sequences.

```python
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Define paths
raw_videos_path = '../data/raw/'
keypoints_path = '../data/processed/keypoints/'
skeletons_path = '../data/processed/skeletons/'

# Function to display a video frame
def display_frame(frame):
    plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()

# Example: Display a frame from a video
video_file = os.path.join(raw_videos_path, 'example_video.mp4')
cap = cv2.VideoCapture(video_file)
ret, frame = cap.read()
if ret:
    display_frame(frame)
cap.release()

# Further analysis can be added here, such as visualizing keypoints and skeletons.
```
